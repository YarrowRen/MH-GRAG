```python
import pandas as pd

# 确保 published_at 列是日期时间格式
corpus_df['published_at'] = pd.to_datetime(corpus_df['published_at'])

# 获取最早的时间
start_time = corpus_df['published_at'].min()

# 获取最晚的时间
end_time = corpus_df['published_at'].max()

print(f"起始时间：{start_time}")
print(f"终止时间：{end_time}")
```

```python
from transformers import BertTokenizer

# 加载 BERT 分词器
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# 计算每条文本的 token 数量
def count_tokens(text):
    return len(tokenizer.tokenize(text))

# 对 body 列应用分词器，并计算总 token 数量
corpus_df['token_count'] = corpus_df['body'].apply(count_tokens)
total_tokens = corpus_df['token_count'].sum()

print(f"总 token 数量：{total_tokens}")
```

```python
from transformers import BertTokenizer
import pandas as pd

# 加载 BERT 分词器
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# 将 'body' 列中的文本进行分词并按 300 tokens 进行切分
def split_into_chunks(text, chunk_size=300):
    # 将文本分词
    tokens = tokenizer.tokenize(text)
    
    # 按照 chunk_size 切分 tokens
    chunks = [' '.join(tokens[i:i + chunk_size]) for i in range(0, len(tokens), chunk_size)]
    
    # 检查是否存在不足 chunk_size 的 chunk
    if len(chunks) > 1 and len(tokenizer.tokenize(chunks[-1])) < chunk_size:
        # 将最后一个 chunk 合并到倒数第二个 chunk
        chunks[-2] = chunks[-2] + ' ' + chunks[-1]
        # 删除最后一个 chunk
        chunks.pop()
    
    return chunks

# 创建一个新的列，将 'body' 列的文本按 300 tokens 切分
corpus_df['body_chunks'] = corpus_df['body'].apply(lambda x: split_into_chunks(x))

# 展开每个文档的 chunks，使每个 chunk 成为一行，并保留原有的列信息
expanded_df = corpus_df.explode('body_chunks')

# 选择需要保留的列并重命名 'body_chunks' 为 'chunk'
corpus_chunks_df = expanded_df[['category', 'author', 'published_at', 'title', 'source', 'body_chunks']].rename(columns={'body_chunks': 'chunk'})
```

```python
# 为每一行生成顺序 ID
corpus_chunks_df = corpus_chunks_df.reset_index(drop=True)  # 重置索引
corpus_chunks_df['id'] = corpus_chunks_df.index  # 使用重置后的索引作为 ID
# 生成顺序 ID，从 1 开始
corpus_chunks_df['id'] = range(1, len(corpus_chunks_df) + 1)

# 将 'id' 列放到第一列
columns = ['id'] + [col for col in corpus_chunks_df.columns if col != 'id']
corpus_chunks_df = corpus_chunks_df[columns]
corpus_chunks_df
```
