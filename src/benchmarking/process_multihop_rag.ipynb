```python
from transformers import BertTokenizer
import pandas as pd

# 加载 BERT 分词器
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# 将 'body' 列中的文本进行分词并按 300 tokens 进行切分，确保句子不被切分
def split_into_chunks(text, chunk_size=300, min_last_chunk_size=150):
    # 将文本分词
    tokens = tokenizer.tokenize(text)

    # 保存每个 token 对应的原始文本标点符号（句子的结束位置）
    sentence_endings = {'.', '!', '?'}
    
    chunks = []
    current_chunk = []
    
    for i, token in enumerate(tokens):
        current_chunk.append(token)

        # 每 chunk_size 个 token 检查是否可以在句子结束符号处切分
        if len(current_chunk) >= chunk_size:
            if token in sentence_endings or (i + 1 < len(tokens) and tokens[i + 1][0] == '##'):
                # 在句子结束符或下一个 token 是子词时进行切分
                chunks.append(current_chunk)
                current_chunk = []

    # 如果剩余 token 不为空，加入最后的 chunk
    if current_chunk:
        chunks.append(current_chunk)
    
    # 如果最后一个 chunk 小于指定的最小大小（如 100 tokens），则将其合并到倒数第二个 chunk
    if len(chunks) > 1 and len(chunks[-1]) < min_last_chunk_size:
        chunks[-2].extend(chunks[-1])  # 将最后一个 chunk 合并到倒数第二个
        chunks.pop()  # 删除最后一个 chunk
    
    # 将每个 chunk 转为字符串，并去除 '##'
    chunks = [tokenizer.convert_tokens_to_string(chunk) for chunk in chunks]
    
    return chunks

# 创建一个新的列，将 'body' 列的文本按 300 tokens 切分
corpus_df['body_chunks'] = corpus_df['body'].apply(lambda x: split_into_chunks(x))

# 展开每个文档的 chunks，使每个 chunk 成为一行，并保留原有的列信息
expanded_df = corpus_df.explode('body_chunks')

# 为每一行生成顺序 ID
expanded_df = expanded_df.reset_index(drop=True)  # 重置索引
expanded_df['id'] = expanded_df.index  # 使用重置后的索引作为 ID

# 选择需要保留的列并重命名 'body_chunks' 为 'chunk'
corpus_chunks_df = expanded_df[['id', 'category', 'author', 'published_at', 'title', 'source', 'body_chunks']].rename(columns={'body_chunks': 'chunk'})
corpus_chunks_df
```
