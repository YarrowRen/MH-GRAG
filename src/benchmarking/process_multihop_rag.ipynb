```python
from transformers import BertTokenizer
import pandas as pd

# Âä†ËΩΩ BERT ÂàÜËØçÂô®
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Â∞Ü 'body' Âàó‰∏≠ÁöÑÊñáÊú¨ËøõË°åÂàÜËØçÂπ∂Êåâ 300 tokens ËøõË°åÂàáÂàÜ
def split_into_chunks(text, chunk_size=300):
    # Â∞ÜÊñáÊú¨ÂàÜËØç
    tokens = tokenizer.tokenize(text)
    
    # ÊåâÁÖß chunk_size ÂàáÂàÜ tokens
    chunks = [' '.join(tokens[i:i + chunk_size]) for i in range(0, len(tokens), chunk_size)]
    
    # Ê£ÄÊü•ÊòØÂê¶Â≠òÂú®‰∏çË∂≥ chunk_size ÁöÑ chunk
    if len(chunks) > 1 and len(tokenizer.tokenize(chunks[-1])) < chunk_size:
        # Â∞ÜÊúÄÂêé‰∏Ä‰∏™ chunk ÂêàÂπ∂Âà∞ÂÄíÊï∞Á¨¨‰∫å‰∏™ chunk
        chunks[-2] = chunks[-2] + ' ' + chunks[-1]
        # Âà†Èô§ÊúÄÂêé‰∏Ä‰∏™ chunk
        chunks.pop()
    
    return chunks

# ÂàõÂª∫‰∏Ä‰∏™Êñ∞ÁöÑÂàóÔºåÂ∞Ü 'body' ÂàóÁöÑÊñáÊú¨Êåâ 300 tokens ÂàáÂàÜ
corpus_df['body_chunks'] = corpus_df['body'].apply(lambda x: split_into_chunks(x))

# Â±ïÂºÄÊØè‰∏™ÊñáÊ°£ÁöÑ chunksÔºå‰ΩøÊØè‰∏™ chunk Êàê‰∏∫‰∏ÄË°åÔºåÂπ∂‰øùÁïôÂéüÊúâÁöÑÂàó‰ø°ÊÅØ
expanded_df = corpus_df.explode('body_chunks')

# ÈÄâÊã©ÈúÄË¶Å‰øùÁïôÁöÑÂàóÂπ∂ÈáçÂëΩÂêç 'body_chunks' ‰∏∫ 'chunk'
corpus_chunks_df = expanded_df[['category', 'author', 'published_at', 'title', 'source', 'body_chunks']].rename(columns={'body_chunks': 'chunk'})
```

```python
# ‰∏∫ÊØè‰∏ÄË°åÁîüÊàêÈ°∫Â∫è ID
corpus_chunks_df = corpus_chunks_df.reset_index(drop=True)  # ÈáçÁΩÆÁ¥¢Âºï
corpus_chunks_df['id'] = corpus_chunks_df.index  # ‰ΩøÁî®ÈáçÁΩÆÂêéÁöÑÁ¥¢Âºï‰Ωú‰∏∫ ID
# ÁîüÊàêÈ°∫Â∫è IDÔºå‰ªé 1 ÂºÄÂßã
corpus_chunks_df['id'] = range(1, len(corpus_chunks_df) + 1)

# Â∞Ü 'id' ÂàóÊîæÂà∞Á¨¨‰∏ÄÂàó
columns = ['id'] + [col for col in corpus_chunks_df.columns if col != 'id']
corpus_chunks_df = corpus_chunks_df[columns]
corpus_chunks_df
```
