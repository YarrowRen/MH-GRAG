```python
import torch
import pandas as pd
import numpy as np
from torch_geometric.data import Data
from sklearn.neighbors import NearestNeighbors

# 创建原始图，节点为entity_id，边为source_entity_id和target_entity_id
def create_graph_from_df(entities_df, relationships_df):
    # 将entity_id映射到连续的节点索引
    entity_id_to_index = {entity_id: idx for idx, entity_id in enumerate(entities_df['entity_id'])}
    
    # 将source和target的entity_id转换为索引
    source_indices = relationships_df['source_entity_id'].map(entity_id_to_index).values
    target_indices = relationships_df['target_entity_id'].map(entity_id_to_index).values
    
    # 构建PyTorch Geometric格式的图结构（edge_index）
    edge_index = torch.tensor([source_indices, target_indices], dtype=torch.long)
    
    # 创建图数据对象（不带特征和边权重）
    graph_data = Data(edge_index=edge_index)
    
    return graph_data, entity_id_to_index

# 基于embedding生成K近邻图，并为每条边分配固定的权重
def generate_knn_graph_from_embeddings(embedding, k=10, fixed_weight=1.0):
    embedding = embedding.cpu().numpy()  # 确保在CPU上进行操作
    nbrs = NearestNeighbors(n_neighbors=k, algorithm='auto').fit(embedding)
    distances, indices = nbrs.kneighbors(embedding)

    num_nodes = embedding.shape[0]
    adjacency_matrix = np.zeros((num_nodes, num_nodes))

    # 为K近邻图的每条边分配固定的权重
    for i in range(num_nodes):
        for j in range(1, k):  # 跳过第一个点自己本身
            adjacency_matrix[i, indices[i][j]] = fixed_weight  # 设置固定权重

    return adjacency_matrix

# 将K近邻生成的邻接矩阵转换为PyTorch Geometric的edge_index和edge_weight
def knn_to_edge_index(knn_graph):
    sources, targets = knn_graph.nonzero()  # 获取非零元素索引
    weights = knn_graph[sources, targets]   # 获取对应的权重
    edge_index_knn = torch.tensor([sources, targets], dtype=torch.long)
    edge_weight_knn = torch.tensor(weights, dtype=torch.float)
    return edge_index_knn, edge_weight_knn

# 原始图和每个头的K近邻图分别融合
def combine_graphs_separately(original_graph, knn_graphs):
    combined_graphs = []
    
    # 原始图的边索引和边权重（初始假设为1）
    edge_index_original = original_graph.edge_index
    edge_weight_original = torch.ones(original_graph.edge_index.shape[1])  # 固定权重为1

    for idx, knn_graph in enumerate(knn_graphs):
        # 将K近邻图转换为edge_index和edge_weight格式
        edge_index_knn, edge_weight_knn = knn_to_edge_index(knn_graph)

        # 分别将原始图与每个头的K近邻图融合
        edge_index_combined = torch.cat([edge_index_original, edge_index_knn], dim=1)
        edge_weight_combined = torch.cat([edge_weight_original, edge_weight_knn], dim=0)

        print(f"Head {idx}: Combined edge index shape: {edge_index_combined.shape}")
        print(f"Head {idx}: Combined edge weights shape: {edge_weight_combined.shape}")
        
        # 存储融合后的图
        combined_graphs.append((edge_index_combined, edge_weight_combined))
    
    return combined_graphs


# 生成原始图
graph_data, entity_id_to_index = create_graph_from_df(filtered_final_entities, filtered_final_relationships)


# 生成每个头的K近邻图，固定边权重为1.0
knn_graphs = []
for embedding in embeddings_list:
    knn_graph = generate_knn_graph_from_embeddings(embedding, k=2, fixed_weight=1.0)
    knn_graphs.append(knn_graph)

# 分别融合原始图和每个头的K近邻图
combined_graphs = combine_graphs_separately(graph_data, knn_graphs)

# 输出最终结果
for idx, (combined_edge_index, combined_edge_weights) in enumerate(combined_graphs):
    print(f"Final combined graph for Head {idx}:")
    print(f"Edge index: {combined_edge_index}")
    print(f"Edge weights: {combined_edge_weights}")
```

```python
import torch
import numpy as np
import igraph as ig
import leidenalg
from sklearn.metrics import normalized_mutual_info_score
import pandas as pd

# 将PyTorch Geometric格式的图数据转换为igraph格式
def convert_to_igraph(edge_index):
    num_nodes = int(edge_index.max().item() + 1)
    edges = edge_index.t().tolist()  # 转换为边的列表
    graph = ig.Graph(edges=edges, directed=False)
    graph.add_vertices(num_nodes - len(graph.vs))  # 添加缺失的顶点
    return graph

# 使用Leiden算法执行社区检测
def run_leiden(graph, edge_weights=None):
    partition = leidenalg.find_partition(graph, leidenalg.ModularityVertexPartition, weights=edge_weights)
    return partition

# 计算互信息
def compute_mutual_info(partitions):
    num_partitions = len(partitions)
    mutual_info_matrix = np.zeros((num_partitions, num_partitions))
    
    for i in range(num_partitions):
        for j in range(i, num_partitions):
            if i == j:
                mutual_info_matrix[i, j] = 1.0
            else:
                mutual_info = normalized_mutual_info_score(partitions[i], partitions[j])
                mutual_info_matrix[i, j] = mutual_info
                mutual_info_matrix[j, i] = mutual_info
    
    return mutual_info_matrix

# 创建一个函数来运行整个流程并存储聚类结果
def analyze_graphs(original_graph, combined_graphs, filtered_final_entities):
    partitions = []
    modularities = []
    cluster_counts = []
    
    # 对原始图执行Leiden算法
    print("\nRunning Leiden for Original Graph")
    original_partition = run_leiden(original_graph)
    original_modularity = original_partition.modularity
    original_cluster_count = len(original_partition)
    print(f"Original Graph: Cluster Count = {original_cluster_count}, Modularity = {original_modularity}")
    
    # 保存原始图的聚类信息
    partitions.append(original_partition.membership)
    modularities.append(original_modularity)
    cluster_counts.append(original_cluster_count)
    
    # 添加原始图的聚类结果到 filtered_final_entities 中
    filtered_final_entities['cluster_original'] = original_partition.membership
    
    # 对每个头的图执行Leiden算法
    for idx, (edge_index, edge_weights) in enumerate(combined_graphs):
        print(f"\nRunning Leiden for Head {idx}")
        
        # 转换为 igraph 格式
        graph = convert_to_igraph(edge_index)
        
        # 执行Leiden算法
        partition = run_leiden(graph, edge_weights.tolist())
        
        # 获取社区信息
        modularity = partition.modularity
        cluster_count = len(partition)
        print(f"Head {idx}: Cluster Count = {cluster_count}, Modularity = {modularity}")
        
        # 存储每个头的聚类结果、模块度、簇数
        partitions.append(partition.membership)
        modularities.append(modularity)
        cluster_counts.append(cluster_count)
        
        # 添加每个头的聚类结果到 filtered_final_entities 中
        filtered_final_entities[f'cluster_head_{idx}'] = partition.membership
    
    # 计算头之间以及与原始图的互信息
    mutual_info_matrix = compute_mutual_info(partitions)
    print("\nMutual Information between heads and original graph:")
    print(mutual_info_matrix)

    return modularities, cluster_counts, mutual_info_matrix, filtered_final_entities

# 将 source 和 target 列表转换为 PyTorch 张量 (2xN)
source_nodes = torch.tensor(filtered_final_relationships['source_entity_id'], dtype=torch.long)
target_nodes = torch.tensor(filtered_final_relationships['target_entity_id'], dtype=torch.long)

# 构建 original_edge_index (2xN)
original_edge_index = torch.stack([source_nodes, target_nodes], dim=0)

# 假设 original_edge_index 是原始图的edge_index，combined_graphs 已经初始化
original_graph = convert_to_igraph(original_edge_index)

# 执行流程并存储聚类结果
modularities, cluster_counts, mutual_info_matrix, updated_filtered_final_entities = analyze_graphs(
    original_graph, combined_graphs, filtered_final_entities)

# 输出最终结果
for idx in range(len(modularities)):
    if idx == 0:
        print(f"Original Graph: Modularity = {modularities[idx]}, Cluster Count = {cluster_counts[idx]}")
    else:
        print(f"Head {idx-1}: Modularity = {modularities[idx]}, Cluster Count = {cluster_counts[idx]}")

print("\nMutual Information between different heads and the original graph:")
print(mutual_info_matrix)

# 打印更新后的 filtered_final_entities
updated_filtered_final_entities
```
