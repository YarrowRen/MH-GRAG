```python
from src.utils.llm_helpers import call_llm_api
from tqdm import tqdm  # 引入 tqdm 库

# 假设 person_entities 是一个 DataFrame 对象，其中包含 'person_name' 和 'description' 列
import pandas as pd

# 创建一个新的列 'gender'，初始化为 None
person_entities['gender'] = None

# 使用 tqdm 包裹 person_entities.iterrows()，并设置描述信息显示进度
for index, person in tqdm(person_entities.iterrows(), total=person_entities.shape[0], desc="Processing Persons"):
    # 从每一行提取 person_name 和 description
    person_name = person['entity_name']
    description = person['description']
    
    # 根据输入的 person_name 和 description 生成 message_content
    message_content = f"""
    Please determine the gender of the person based on the information provided below:

    Person Name: {person_name}
    Description: {description}

    Possible answers: male, female, other, unknown

    Please return only one of the following values: male, female, other, or unknown.
    """
    
    # 构造 query_template
    query_template = [
        {"role": "user", "content": message_content}
    ]
    
    # 调用 LLM API 获取性别判断结果
    answer = call_llm_api(query_template)
    
    # 将性别结果存入 'gender' 列
    person_entities.at[index, 'gender'] = answer

# 查看结果
person_entities
```
