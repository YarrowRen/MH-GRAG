```python
final_relationships=load_df_from_csv_with_embedding(f"/home/bren901/coding/MH-GRAG/export/relationships_douqtaig.csv",['relationship_embedding'])
final_relationships
```

```python
# 筛选 final_entities 中 corpus_id < 2000 的数据
filtered_final_entities = final_entities[final_entities['corpus_id'] < 2000]

# 筛选 final_relationships 中 source_entity_id 和 target_entity_id 对应的 corpus_id 都 < 2000 的数据
filtered_final_relationships = final_relationships[
    (final_relationships['source_entity_id'].isin(filtered_final_entities['entity_id'])) &
    (final_relationships['target_entity_id'].isin(filtered_final_entities['entity_id']))
]

filtered_final_entities
```

```python
from src.gnn_clustering.data_loader import load_custom_data
from src.gnn_clustering.utils import get_device, get_dense_adj, communities_to_labels
from src.gnn_clustering.multihead_clustering import get_initial_leiden_results, get_clustering_results, convert_to_igraph
from src.utils.graph_utils import create_graph_from_df, generate_knn_graph_from_embeddings, combine_graphs_separately
from src.gnn_clustering.model import get_multi_head_model
from src.gnn_clustering.train import train_model_multi_head
from src.gnn_clustering.evaluate import get_embeddings_list
import torch

# 设置设备
device = get_device()

# 数据加载
data = load_custom_data(filtered_final_entities, filtered_final_relationships)
data = data.to(device)

# 训练模型
num_heads=3
multi_head_model = get_multi_head_model(data=data, device=device, num_heads=num_heads)
optimizer = torch.optim.Adam(multi_head_model.parameters(), lr=0.01)
adj = get_dense_adj(data.edge_index, device=device)

trained_model = train_model_multi_head(multi_head_model, data, adj, optimizer, num_heads)
embeddings_list_after_training = get_embeddings_list(trained_model, data, device)

graph_data, entity_id_to_index = create_graph_from_df(filtered_final_entities, filtered_final_relationships)
knn_graphs = []
for embedding in embeddings_list_after_training:
    knn_graph = generate_knn_graph_from_embeddings(embedding, k=2, fixed_weight=1.0)
    knn_graphs.append(knn_graph)

combined_graphs = combine_graphs_separately(graph_data, knn_graphs)

source_nodes = torch.tensor(filtered_final_relationships['source_entity_id'], dtype=torch.long)
target_nodes = torch.tensor(filtered_final_relationships['target_entity_id'], dtype=torch.long)

original_edge_index = torch.stack([source_nodes, target_nodes], dim=0)
original_graph = convert_to_igraph(original_edge_index)


# 执行测试
merge_leiden_results = get_initial_leiden_results(data=data, device=device)
multi_head_leiden_results = get_clustering_results(combined_graphs)

merge_communities_leiden = merge_leiden_results[0][0]
cluster_labels_merge = communities_to_labels(merge_communities_leiden, data.num_nodes)
filtered_final_entities['cluster_original'] = cluster_labels_merge

# 将多头聚类结果添加到DataFrame
for head_index, clustering_result in enumerate(multi_head_leiden_results):
    cluster_labels_multihead = communities_to_labels(clustering_result, data.num_nodes)
    filtered_final_entities[f'cluster_head_{head_index}'] = cluster_labels_multihead

# 打印结果确认
filtered_final_entities

```
